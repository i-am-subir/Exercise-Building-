```python
# these are the library we are going to use for this dataset.
# import other libraries further if needed..

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# importing the CSV data file..

df = pd.read_csv('sales_data.csv',parse_dates=['Date'])
```

let the play begin.

```python
df['Year'].value_counts()    #  this insight is plotted below..
```

```python
# Checking the sales over the year.

sns.set_style('whitegrid')
sns.countplot(df['Year'])
plt.title('Sales over the Year')
```

```python
# checking the Customer age density or variation

sns.distplot(df['Customer_Age'])
plt.title('Customer age density distribution chart')
```

```python
# Here we checking the the gender ratio of the customers in this dataset..


df['Customer_Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df.head()
```

```python
df['Age_Group'].value_counts()
```

```python
# checking the customer age group and customer age alignment are correct or not....
# As a data scientist you need to varify the given data parameters always true...for accurate insights


sns.barplot(x='Age_Group',y='Customer_Age',data=df,hue='Customer_Gender')
```

```python
# Customer age mean 

plt.figure(figsize=(15,3))
sns.boxplot(df['Customer_Age'])

plt.title('Customer Age Distribution')
```

```python
df['State'].value_counts()  # lets plot this graph below
```

```python
plt.figure(figsize=(15,15))
sns.set_style('whitegrid')

ax = sns.countplot(y=df['State'])
ax = ax.bar_label(ax.containers[0])

plt.title('State wise Distribution')
```

```python
df.head()
```

```python
df['Sub_Category'].value_counts()
```

```python
df['Product_Category'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Items In Product Category')
```

```python
# Grouping Product Category Wise.
cv = df.groupby('Product_Category')


profit_category = cv['Profit'].sum().index
profit_Values = cv['Profit'].sum().values


plt.pie(profit_Values,labels=profit_category,autopct='%1.1f%%')

plt.title('Profit Percentage in Product Category')
```

```python
plt.figure(figsize=(20,4))
sns.boxplot(x='Profit',y='Product_Category',data=df)

plt.title('Profit Distribution in Product Category')
```

```python
plt.figure(figsize=(12,3))
sns.set_style('whitegrid')

sns.scatterplot(y='Country',x='Revenue',data=df,hue='Product_Category',s=20)

plt.title('Country Wise Revenue Distribution')
```

```python
plt.figure(figsize=(20,15))
sns.set_style('whitegrid')

sns.scatterplot(y='State',x='Revenue',data=df,hue='Product_Category',s=20)

plt.title('State v/s Revenue Distribution')
```
